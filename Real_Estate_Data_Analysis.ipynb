```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample real estate dataset
data = {
    "Location": ["East Zone", "City Center", "East Zone", "Highway Side",
                 "City Center", "East Zone", "Highway Side", "City Center"],
    "Property_Type": ["Apartment", "Apartment", "Apartment", "Villa",
                      "Apartment", "Apartment", "Villa", "Apartment"],
    "Size_sqft": [900, 1100, 950, 2000, 1200, 1000, 1800, 1050],
    "Bedrooms": [2, 3, 2, 4, 3, 2, 3, 2],
    "Parking": [1, 1, 0, 2, 1, 1, 2, 1],
    "Price_Lakhs": [55, 85, 60, 150, 90, 65, 140, 80]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display dataset
print("Real Estate Dataset")
display(df)

# Basic statistics
print("\nStatistical Summary")
display(df.describe())

# Average price by location
avg_price_location = df.groupby("Location")["Price_Lakhs"].mean()
print("\nAverage Price by Location")
display(avg_price_location)

# Visualization: Average price by location
plt.figure()
avg_price_location.plot(kind="bar")
plt.title("Average Property Price by Location")
plt.xlabel("Location")
plt.ylabel("Price (Lakhs)")
plt.show()

# Visualization: Price vs Size
plt.figure()
plt.scatter(df["Size_sqft"], df["Price_Lakhs"])
plt.title("Property Size vs Price")
plt.xlabel("Size (sqft)")
plt.ylabel("Price (Lakhs)")
plt.show()

# Save dataset for GitHub upload
df.to_csv("real_estate_data.csv", index=False)
print("\nDataset saved as real_estate_data.csv")
```
